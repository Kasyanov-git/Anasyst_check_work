```python
   import sqlite3
```

```python
connection = sqlite3.connect("Y:/test.db")
cursor = connection.cursor()
```

```python
cursor.execute("""SELECT * FROM history""")

results = cursor.fetchall()

for row in results:
    print(row)
```

```python
cursor.execute("""SELECT SUBSTR(issue_key, 1, 1) AS group_name, ROUND(AVG(minutes_in_status)/60.0, 2) AS avg_time_in_hours
FROM history
WHERE status = 'Open'
GROUP BY group_name;""")

results = cursor.fetchall()

for row in results:
    print(row)
```

```python
cursor.execute("""SELECT issue_key, status, strftime('%Y-%m-%d %H:%M:%S', datetime(started_at/1000, 'unixepoch')) AS creation_time
FROM history
WHERE status NOT IN ('Closed', 'Resolved')
AND (ended_at IS NULL OR ended_at > strftime('%s', 'now')*1000)
ORDER BY creation_time;""")

results = cursor.fetchall()

for row in results:
    print(row)
```
